```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/chronic-kidney-disease.csv')
```

```python
data.head()
```

```python
cols = list(data.columns)
bp, classes = cols.index('Bp'), cols.index('Class')
cols[bp], cols[classes] = cols[classes], cols[bp]
data = data[cols]
```

```python
data.head()
```

```python
# remove rbc, htn (nominal)

data.drop(['Rbc','Htn'], axis=1, inplace=True)
data.head()
```

https://www.geeksforgeeks.org/normalize-a-column-in-pandas/

```python
# normalize Bu, Sc, Sod, Pot, Hemo, Wbcc, Rbcc, Bp

df_min_max_scaled = data.copy()
  
# apply normalization techniques by Column 1
column = ['Bu', 'Sc', 'Sod', 'Pot', 'Hemo', 'Wbcc', 'Rbcc', 'Bp']
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  
# view normalized data
display(df_min_max_scaled)
```

```python
df_min_max_scaled.to_csv('data/chronic-kidney-disease-updated.csv', index=False, header=False, sep=' ')
```
